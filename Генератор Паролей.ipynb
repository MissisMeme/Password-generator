```python
import random as r 

digits = '0123456789'
lowercase_letters = 'abcdefghijklmnopqrstuvwxyz'
uppercase_letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
punctuation = '!#$%&*+-=?@^_'
ambiguous = 'il1Lo0O'
chars = []


def is_valid_number(n):
    while not n.isdigit() and int(n) != 0:
        n = input('Введите число больше 0. Попробуйте еще раз. Сколько паролей требуется создать?')
    if n.isdigit() and int(n) > 0:
        n = int(n)
        return n    

def is_valid_lenth(n):
    while not n.isdigit() and int(n) != 0:
        n = input('Введите число больше 0. Длина одного пароля:')
    if n.isdigit() and int(n) > 0:
        n = int(n)
        return n  

def is_valid_answer(a):
    a = a.lower()
    while a != 'да' and a != 'нет':
        a = input('Я вас не понимаю. Введите ДА или НЕТ') 
    if a == 'да' or a == 'нет':        
        return a       
    

number = input('Сколько паролей вам требуется создать? (Введите число)')
number = is_valid_number(number)

lenth = input('Длина одного пароля: (Введите число)')
lenth = is_valid_lenth(lenth)

answer_lowercase = input('Пароль должен включать строчные буквы? (Да/Нет)')
answer_lowercase = is_valid_answer(answer_lowercase)

answer_uppercase = input('Пароль должен включать прописные буквы? (Да/Нет)')
answer_uppercase = is_valid_answer(answer_uppercase)

answer_digits = input('Включать цифры?(Да/Нет)')
answer_digits = is_valid_answer(answer_digits)

answer_punctuation = input('Включать ли символы? (Да/Нет)')
answer_punctuation = is_valid_answer(answer_punctuation)

exclude = input('Вы хотите исключить неоднозначные символы? (Да/Нет)')
exclude = is_valid_answer(exclude)

if answer_lowercase == 'да':
    chars.append(lowercase_letters)
if answer_uppercase == 'да':
    chars.append(uppercase_letters)
if answer_digits == 'да':
    chars.append(digits)
if answer_punctuation == 'да':
    chars.append(punctuation)
if exclude == 'да':
    for c in chars:
        temp = c
        for k in temp:
            if k in 'il1Lo0O':
                ind = temp.index(k)
                temp = temp[0:ind] + temp[ind + 1:]
        chars[chars.index(c)] = temp
                

def generate_password(lent, chr):   
    # Делим нужное кол-во символов на кол-во элементов в списке, из к-го будем брать символы для пароля.           
    ran = lent // len(chr)   
    # Ищем остаток от деления нужного кол-ва символов на кол-во элементов в списке.                  
    mod = lent % len(chr)    
    password = ''
    
    if ran:
        for i in range(ran):
            r.shuffle(chr)
            for j in chr:
                password += r.choice(j)
        if mod:
            r.shuffle(chr)
            chr = r.sample(chr, mod)
            for k in chr:
                password += r.choice(k)

    return password

for f in range(number):
    p_word = generate_password(lenth, chars)
    print(p_word)           

```
